### Graph: x vs y

Lets first draw the graph which will plot the path of the pendulum in the xy plane

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plott(lamda, Omega, length, t):
    
    L = length
    R = Omega
    lamda = lamda
    g=9.8 
    # we will take the value of C1 and C2 to be equal to 1
    k_1=1
    k_2=1  
    
    a=np.exp(1j*(np.sqrt(g/L)*t))
    b=np.exp(1j*((-1)*np.sqrt(g/L)*t))
    c=np.exp(1j*(R*np.sin(lamda)*(-1)*t))

    u=(k_1*a+k_2*b)*c
    
    ## separating the real part 
    x=u.real
    y=u.imag
    
    plt.plot(x,y,'g')
    plt.grid()
```

# Foucalt pendulum in the setting of flat earth 

```python
def plott_flat(r, t):
    
#     L = length
    R = 100
    G=13
    sigma= 0.1
    # we will take the value of C1 and C2 to be equal to 1
    k_1=1
    k_2=1  
    
    K_r = G*sigma*((np.pi/R) + (12*np.pi*(r**2)/36*(R**2) ))
    
#     a=np.exp(1j*(np.sqrt(g/L)*t))
    b=np.exp(1j*((-1)*K_r*t))
    c=np.exp(1j*(K_r*t))

    x=(k_1*b+k_2*c)
    y=(2*b+4*c)
#     ## separating the real part 
#     x=u.real
#     y=u.imag
    
    plt.plot(x,y,'g')
    plt.grid()
```

# Walking Backwards from Motion of Pendulum to Latitude

```python
def angle_btw(a,b,c):
    '''
    This will find the angle abc by using the cosine law
    '''
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    ab = a - b
    cb = c - b
    
    cosine_angle  = np.dot(ab, cb) / (np.linalg.norm(ab) * np.linalg.norm(cb))
    angle = np.arccos(cosine_angle)

    return np.degrees(angle)
    
```

```python
def angle_all(a):
    anglee = []
    for i in range(len(a)):
        anglee.append(angle_btw(a[i][0],a[i][1],a[i][2]))
        
    return anglee
```

## Curve Fit

The above graph seems to follow a trend so lets try curve fitting in order to obtain the best fit

```python
from scipy.optimize import curve_fit

def test_fit(x,a,b,c):
    return a*np.exp(b*x) + c


params1, cov = curve_fit(test_fit,petal_angle_rel,lat_deg)
```

```python
## plotting the curve
x = np.arange(0,1.1,0.05)
y = params1[0]*np.exp(params1[1]*x) + params1[2]
plt.plot(x,y, 'r')
plt.plot(petal_angle_rel , lat_deg, 'o-')
plt.ylabel('Latitude angle')
plt.xlabel('Petal Angle(relative)')
plt.grid()
plt.show()
```
